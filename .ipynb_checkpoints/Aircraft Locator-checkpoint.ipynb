```python
to_search = 25
labels_per = 4
```

```python
class Model:
    def __init__(self):
        self.model = self.build_model()
        print(self.model.summary())
    
    def build_model(self):
        model = tf.keras.Sequential([
            tf.keras.layers.InputLayer((640,640,3)),
            tf.keras.layers.Conv2D(filters = 64, kernel_size = 7),
            tf.keras.layers.MaxPool2D(),
            tf.keras.layers.Conv2D(filters = 192, kernel_size = 3),
            tf.keras.layers.MaxPool2D(),
            tf.keras.layers.Conv2D(filters = 128, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 256, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 256, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 512, kernel_size = 3),
            tf.keras.layers.MaxPool2D(),
            tf.keras.layers.Conv2D(filters = 256, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 512, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 512, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.MaxPool2D(),
            tf.keras.layers.Conv2D(filters = 256, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 512, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 512, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.MaxPool2D(),
            tf.keras.layers.Conv2D(filters = 512, kernel_size = 1),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.Conv2D(filters = 1024, kernel_size = 3),
            tf.keras.layers.Flatten(),
            tf.keras.layers.Dense(4096, activation=tf.keras.layers.LeakyReLU(alpha=0.01)),
            tf.keras.layers.Dense((to_search*labels_per), activation='tanh'),
            tf.keras.layers.Reshape((-1,4))
        ])
        opt = tf.keras.optimizers.Adam(learning_rate=0.001)
        model.compile(optimizer=opt, 
              loss='mean_squared_logarithmic_error', metrics=['accuracy'])

        return model
        
```

```python
def sliding_window(image, kernel_size: tuple([int, int]), step: tuple([int, int])):
    for y in range(0, image.shape[0], step[1]):
        for x in range(0, image.shape[1], step[0]):
            yield (x, y, image[y: y+kernel_size[1], x: x+kernel_size[0]])
```
